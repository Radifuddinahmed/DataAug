```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.cm import get_cmap

var='depth'
df = pd.read_csv(f'hp_{var}.csv')
# Read data from CSV file
#df = pd.read_csv('width_new_model_comparison_results_20231214192952.csv')  # Replace with the actual path to your CSV file

# Extract model names and error values
models = df['Model'].tolist()
mae_values = df['MAE'].tolist()
rmse_values = df['RMSE'].tolist()
rrse_values = df['RRSE'].tolist()
rae_values = df['RAE'].tolist()
correlation_values = df['Correlation Coefficient'].tolist()
#correlation_values = df['R-squared'].tolist()


# Define colors for each model
#colors = plt.cm.viridis(np.linspace(0, 1, len(models)))

cmap = get_cmap('Set3')
colors = cmap(np.linspace(0, 1, len(models)))

# Create subplots
fig, axes = plt.subplots(nrows=1, ncols=5, figsize=(20, 5))

# Plot Mean Absolute Error
for i, color in zip(range(len(models)), colors):
    axes[0].bar(i, mae_values[i], color=color)
axes[0].set_title('Mean Absolute Error (MAE)')
#axes[0].set_ylabel('MAE')
axes[0].set_ylim([0.005,0.01])

# Plot Root Mean Squared Error
for i, color in zip(range(len(models)), colors):
    axes[1].bar(i, rmse_values[i], color=color)
axes[1].set_title('Root Mean Squared Error (RMSE)')
#axes[1].set_ylabel('RMSE')
axes[1].set_ylim([0.024, 0.035])

# Plot Relative Root Squared Error
for i, color in zip(range(len(models)), colors):
    axes[2].bar(i, rrse_values[i], color=color)
axes[2].set_title('Relative Root Squared Error (RRSE)')
#axes[2].set_ylabel('RRSE')
axes[2].set_ylim([0.25, 0.35])

# Plot Relative Absolute Error
for i, color in zip(range(len(models)), colors):
    axes[3].bar(i, rae_values[i], color=color)
axes[3].set_title('Relative Absolute Error (RAE)')
#axes[3].set_ylabel('RAE')
axes[3].set_ylim([0.12, 0.22])

# Plot Correlation Coefficient
for i, color in zip(range(len(models)), colors):
    axes[4].bar(i, correlation_values[i], color=color, label=f'{models[i]}')

axes[4].set_title('Correlation Coefficient')
axes[4].set_ylim([0.55, 0.92])  # Set y-axis range for correlation coefficient
#axes[4].set_ylabel('Correlation')

# Add legend to the last subplot
axes[4].legend(loc='upper center', bbox_to_anchor=(-1.75, 1.25), ncol=6)


# Remove x-axis ticks and labels for all subplots
for ax in axes:
    ax.set_xticks([])
    ax.set_xticklabels([])

# Adjust layout for better spacing
#plt.tight_layout()



# Save the plot as a PNG file
plt.savefig(f'output_plot_{var}.png', bbox_inches='tight')

# Show the plot
plt.show()
```

```python
# example bar chart generator
import matplotlib.pyplot as plt
import numpy as np

# Example error metric values for three models
models = ['Model A', 'Model B', 'Model C']
mae_values = [3.2, 2.8, 3.0]
rmse_values = [4.0, 3.5, 3.8]
rrse_values = [0.15, 0.13, 0.14]
rae_values = [0.10, 0.08, 0.09]  # Example Relative Absolute Error
correlation_values = [0.95, 0.92, 0.94]  # Example Correlation Coefficient

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))

# Plot Mean Absolute Error
axes[0, 0].bar(models, mae_values, color=['blue', 'green', 'orange'])
axes[0, 0].set_title('Mean Absolute Error (MAE)')
axes[0, 0].set_ylabel('MAE')

# Plot Root Mean Squared Error
axes[0, 1].bar(models, rmse_values, color=['blue', 'green', 'orange'])
axes[0, 1].set_title('Root Mean Squared Error (RMSE)')
axes[0, 1].set_ylabel('RMSE')

# Plot Relative Root Squared Error
axes[0, 2].bar(models, rrse_values, color=['blue', 'green', 'orange'])
axes[0, 2].set_title('Relative Root Squared Error (RRSE)')
axes[0, 2].set_ylabel('RRSE')

# Plot Relative Absolute Error
axes[1, 0].bar(models, rae_values, color=['blue', 'green', 'orange'])
axes[1, 0].set_title('Relative Absolute Error (RAE)')
axes[1, 0].set_ylabel('RAE')

# Plot Correlation Coefficient
axes[1, 1].bar(models, correlation_values, color=['blue', 'green', 'orange'])
axes[1, 1].set_title('Correlation Coefficient')
axes[1, 1].set_ylabel('Correlation')

# Remove empty subplot
fig.delaxes(axes[1, 2])

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```
