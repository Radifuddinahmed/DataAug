```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 
from matplotlib.cm import get_cmap

var='Depth'
df = pd.read_csv(f'LPBF_ML_Model_Evaluation_BaseModel_{var}.csv')
df2 = pd.read_csv(f'Poly6{var}_base.csv')

#HP
df3 = pd.read_csv(f'hp_{var}.csv')
df4 = pd.read_csv(f'LPBF_ML_Model_Evaluation_Poly6_{var}.csv')


# Read data from CSV file
#df = pd.read_csv('width_new_model_comparison_results_20231214192952.csv')  # Replace with the actual path to your CSV file

# Extract model names and error values
models = df['Model'].tolist()

#BASE
mae_values = df['MAE'].tolist()
rmse_values = df['RMSE'].tolist()
rrse_values = df['RRSE'].tolist()
rae_values = df['RAE'].tolist()
correlation_values = df['Correlation Coefficient'].tolist()
#correlation_values = df['R-squared'].tolist()

#Poly Base
mae_values2 = df2['MAE'].tolist()
rmse_values2 = df2['RMSE'].tolist()
rrse_values2 = df2['RRSE'].tolist()
rae_values2 = df2['RAE'].tolist()
#correlation_values2 = df2['Correlation Coefficient'].tolist()
correlation_values2 = df2['R-squared'].tolist()


#HP BASE
mae_values3 = df3['MAE'].tolist()
rmse_values3 = df3['RMSE'].tolist()
rrse_values3 = df3['RRSE'].tolist()
rae_values3 = df3['RAE'].tolist()
correlation_values3 = df3['Correlation Coefficient'].tolist()

#HP POLY

mae_values4 = df4['MAE'].tolist()
rmse_values4 = df4['RMSE'].tolist()
rrse_values4 = df4['RRSE'].tolist()
rae_values4 = df4['RAE'].tolist()
correlation_values4 = df4['Correlation Coefficient'].tolist()

models_n = ['GP', 'LR', 'PR', 'SVR', 'KNN','MLP','RF','GB','AB','B','ET']

bar_width = 0.2  # Width of each bar
bar_space=0.02

x = np.arange(len(models))  # X positions for the categories



# Create subplots
fig, axes = plt.subplots(nrows=1, ncols=5, figsize=(25,4))

# Plot Mean Absolute Error
axes[0].bar(x, mae_values, width=bar_width, label='Base', color='orange')
axes[0].bar(x + bar_width + bar_space, mae_values3, width=bar_width, label='HP', color='green')
axes[0].bar(x + 2 * (bar_width + bar_space), mae_values2, width=bar_width, label='Base + Poly F.E.', color='brown')
axes[0].bar(x + 3 * (bar_width + bar_space), mae_values4, width=bar_width, label='HP + Poly F.E.', color='cyan')





axes[0].set_ylim([0,0.06])
#axes[0].set_yscale('log')

    
axes[0].set_title('Mean Absolute Error (MAE)')
#axes[0].set_ylabel('MAE')


# Plot Root Mean Squared Error
axes[1].bar(x, rmse_values, width=bar_width, label='Base', color='orange')
axes[1].bar(x + bar_width + bar_space, rmse_values3, width=bar_width, label='HP', color='green')
axes[1].bar(x + 2 * (bar_width + bar_space), rmse_values2, width=bar_width, label='Base + Poly F.E.', color='brown')
axes[1].bar(x + 3 * (bar_width + bar_space), rmse_values4, width=bar_width, label='HP + Poly F.E.', color='cyan')

#axes[1].set_yscale('log')
#axes[1].set_ylim([0,0.1])

axes[1].set_title('Root Mean Squared Error (RMSE)')
#axes[1].set_ylabel('RMSE')
axes[1].set_ylim([0,0.1])

# Plot Relative Root Squared Error
axes[2].bar(x, rrse_values, width=bar_width, label='Base', color='orange')
axes[2].bar(x + bar_width, rrse_values3, width=bar_width, label='HP', color='green')
axes[2].bar(x + 2 * bar_width, rrse_values2, width=bar_width, label='Base + Poly F.E.', color='brown')
axes[2].bar(x + 3 * bar_width, rrse_values4, width=bar_width, label='HP + Poly F.E.', color='cyan')


#axes[1].set_yscale('log')
axes[2].set_ylim([0,1])

axes[2].set_title('Relative Root Squared Error (RRSE)')
#axes[2].set_ylabel('RRSE')
#axes[2].set_ylim([0.25, 0.35])

# Plot Relative Absolute Error
axes[3].bar(x, rae_values, width=bar_width, label='Base', color='orange')
axes[3].bar(x + bar_width, rae_values3, width=bar_width, label='HP', color='green')
axes[3].bar(x + 2 * bar_width, rae_values2, width=bar_width, label='Base + Poly F.E.', color='brown')
axes[3].bar(x + 3 * bar_width, rae_values4, width=bar_width, label='HP + Poly F.E.', color='cyan')


axes[3].set_ylim([0,1])

axes[3].set_title('Relative Absolute Error (RAE)')
#axes[3].set_ylabel('RAE')
#axes[3].set_ylim([0.12, 0.22])

axes[4].bar(x, correlation_values, width=bar_width, label='Base', color='orange')
axes[4].bar(x + bar_width, correlation_values3, width=bar_width, label='HP', color='green')
axes[4].bar(x + 2 * bar_width, correlation_values2, width=bar_width, label='Base + Poly F.E.', color='brown')
axes[4].bar(x + 3 * bar_width, correlation_values4, width=bar_width, label='HP + Poly F.E.', color='cyan')


axes[4].set_ylim([0,1])

axes[4].set_title('Correlation Coefficient')


# Add legend to the last subplot
axes[3].legend(loc='upper center', bbox_to_anchor=(-.5, 1.18), ncol=6)
for ax in axes:
    #current_xticks = ax.get_xticks()

    # Adjust the x-axis tick positions (e.g., move labels slightly to the right)
    #new_xticks = [tick + 0.1 for tick in current_xticks]
    #ax.set_xticks(new_xticks)
    #ax.set_xticklabels(models_n)
    ax.set_xticks(x,models,rotation=45, ha='right')
    #ax.set_xticklabels(ax.get_xticks(), position=(0, -0.1, 0))
    #ax.get_xticks(models, position=(1, 0))
    #ax.tick_params(axis='x', pad=40)

# Adjust layout for better spacing
#plt.tight_layout()



# Save the plot as a PNG file
plt.savefig(f'output_plot_pm_base_poly_hp_{var}.png', bbox_inches='tight')

# Show the plot
plt.show()
```
